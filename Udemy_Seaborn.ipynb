## Un peu de Matplotlib avant Seaborn
### http://www.matplotlib.org
### https://github.com/matplotlib/matplotlib
### https://matplotlib.org.2.0.2/gallery.html

```python
a=np.linspace(0,10,11)
b=a**2
x=np.arange(0,10)
y=2*x
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Label Y - Axes 1,0')
axes[1][0].set_xlim(2,6)
axes[1][1].plot(a,b)
axes[1][1].set_title('Titre 1,1')
fig.suptitle('Titre au niveau Figure', fontsize=12)
fig.subplots_adjust(wspace=0.8,hspace=0.4)
fig.savefig('new1_subplot',bbox_inches='tight')
# wspace=0.8 signifie 80% d'un espace de figure
```

```python
z = np.linspace(0,11,10)
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
ax.plot(z,z, label = 'X vs X', color='green',linewidth=2)
# Site web pour les couleurs hex color picker
ax.plot(z,z**2, label = 'X vs X^2', color = '#42d4f5',lw=2, marker='s', markersize=10, markerfacecolor='red',
       markeredgewidth=4,markeredgecolor='orange')
# options de type de ligne possible '--','-','-.',':','steps'
ax.plot(z,z**3, label = 'X vs X^3', color='#f5b942',lw=2, ls='-.', marker='o', markersize=6)
#localisation du point infÃ©rieur gauche de la lÃ©gende
ax.legend(loc=(1.1,0.4))

```

## Distribution au sein de catÃ©gories

```python
sns.boxplot(data=df, y='displ', x='size', hue='origin')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
sns.violinplot(data=df, y='displ', x='size', hue='origin')
```

```python
sns.violinplot(data=df, x='size', y='displ', split=True, inner='quartile')
```

```python
plt.figure(figsize=(5,4),dpi=150)
sns.swarmplot(data=df, y='displ', x='size', size=2)
```

```python
sns.boxenplot(data=df, y='displ', x='size', hue='origin')
plt.legend(bbox_to_anchor=(1.05,1))
```

## Diagramme matriciel (carte de frÃ©quentation)
### Heatmap()
### Clustermap()

```python
df2 = pd.read_csv('country_table.csv')
df2 = df2.set_index('Countries')
```

```python
df2.head()
```

```python
sns.heatmap(df2)
```

```python
# Attention l'age est en annÃ©e contrairement Ã  des pourcentages dans les autres colonnes
sns.heatmap(df2.drop('Life expectancy',axis=1),linewidth=0.5,annot=True)
```

```python
plt.figure(dpi=200)
sns.heatmap(df2.drop('Life expectancy',axis=1),linewidth=0.5,annot=True,cmap='viridis')
```

```python
plt.figure(dpi=150)
sns.clustermap(df2.drop('Life expectancy',axis=1),linewidth=0.5,annot=True)
```

```python
# On peut stopper le clustering des colonnes
plt.figure(dpi=150)
sns.clustermap(df2.drop('Life expectancy',axis=1),linewidth=0.5,annot=True,col_cluster=False)
```
